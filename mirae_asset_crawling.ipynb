### 미래에셋 펀드 크롤링

#### 펀드명 크롤링

```python
# 구글 크롬 사용을 위해 webdriver 다운로드 후 불러오기
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import re
import time

# 현재 사용중인 구글 크롬의 버전 확인 후 해당하는 chromedriver.exe 다운로드 및 chromedirver 불러오기
driver = webdriver.Chrome('./chromedriver.exe')

# 창의 크기를 최대화
# driver.maximize_window()

# 데이터를 다운 받을 금융투자협회 링크 열기
mirae_asset_url = "https://investments.miraeasset.com/fund/all.do?searchFundLoc1=1&searchFundKind2=21&searchEtfYn=N&kind=21&location=1"
driver.get(mirae_asset_url)

fund_name_ul = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, 
                                            "//*[@id='contents']/div[2]/div[3]/ul")))

fund_name_li = fund_name_ul.find_elements(By.TAG_NAME, "li")

fund_name_list = []
for name in fund_name_li:
    ul_li = name.text
    ul_li_list = ul_li.split("\n")
    fund_name_list.append(ul_li_list)

# print(fund_name_list)

df = pd.DataFrame()
df["펀드명"] = fund_name_list
df
```

```python
df["펀드명"] = df["펀드명"].astype(str)
```

```python
df["펀드명"] = df["펀드명"].str.replace("[", "")
df["펀드명"] = df["펀드명"].str.replace("]", "")
df["펀드명"] = df["펀드명"].str.replace("'", "")
```

```python
df = pd.read_csv('./csv/미래에셋펀드명.csv')
```

```python
mirae_fund_name_list = df["펀드명"].tolist()
mirae_fund_name_list
```

```python
# 구글 크롬 사용을 위해 webdriver 다운로드 후 불러오기
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# 현재 사용중인 구글 크롬의 버전 확인 후 해당하는 chromedriver.exe 다운로드 및 chromedirver 불러오기
driver = webdriver.Chrome('./chromedriver.exe')

# 창의 크기를 최대화
# driver.maximize_window()

# 데이터를 다운 받을 금융투자협회 링크 열기
mirae_asset_url = "https://investments.miraeasset.com/"
driver.get(mirae_asset_url)

time.sleep(1)

# 팝업창 닫기
close_popup = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, 
                                            "//*[@id='webseminar']/div/div/div[2]/button[1]")))
# "투자상품" 클릭
close_popup.click()

# 상단 탭 "투자상품"의 XPath
fund_goods = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, 
                                            "//*[@id='gnb']/div/ul/li[1]")))
# "투자상품" 클릭
fund_goods.click()

# "펀드투자상품 찾기"의 XPath
find_fund_goods = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, 
                                            "//*[@id='gnb']/div/ul/li[1]/div/ul/li[1]/a")))
# "펀드투자상품 찾기" 클릭
find_fund_goods.click()

time.sleep(2)

for fund_name in mirae_fund_name_list:

    # "상품명 입력"의 XPath
    input_fund_goods = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, 
                                                "//*[@id='searchText']")))
    # "상품명 입력" 클릭
    input_fund_goods.click()
    input_fund_goods.send_keys(Keys.CONTROL + "A")
    # input_fund_goods.send_keys(Keys.DELETE)
    input_fund_goods.send_keys(fund_name)
    # input_fund_goods.send_keys(Keys.ENTER)
    
    # 돋보기 XPath
    mag_btn = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, 
                                                "//*[@id='searchForm']/section/fieldset/div/button[1]")))

    # 돋보기 클릭
    mag_btn.click()
    driver.execute_script("arguments[0].click();", mag_btn) 

    time.sleep(2)
    
    # 검색 펀드명의 첫번째 옵션 XPath
    fund_search_first_option = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, 
                                                "//*[@id='commonFundList']/tr[1]/td[2]/a")))

    # 검색 펀드명의 첫번째 옵션 클릭
    fund_search_first_option.click() 

    # 새 창으로 창 전환
    driver.switch_to.window(driver.window_handles[1])
    driver.get_window_position(driver.window_handles[1])

    # 기간 목록 XPath
    time_list = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, 
                                                "//*[@id='selectViewRate']")))

    # 기간 목록 클릭
    time_list.click() 

    # 3년 XPath
    year_3 = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, 
                                                "//*[@id='selectViewRate']/option[6]")))

    # 3년 클릭
    year_3.click() 

    # "조회" XPath
    search_year_3 = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, 
                                                "//*[@id='rateChart']/div[2]/button")))

    # "조회" 클릭
    search_year_3.click() 

    time.sleep(2)

    # "csv" XPath
    download_csv = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, 
                                                "//*[@id='csv']")))

    # "csv" 클릭
    download_csv.click()
    driver.close()
```
