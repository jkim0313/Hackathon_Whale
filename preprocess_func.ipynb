```python
def merge_preprocess(df1, df2):
    merged_df = pd.merge(df1, df2, how="left", on="펀드명")
    merged_df = merged_df[["펀드명", "수익률 및 순위_x", "수익률 및 순위.2_x", "수익률 및 순위.4_x", "수익률 및 순위.6_x", "수익률 및 순위_y", "수익률 및 순위.2_y", "수익률 및 순위.4_y", "수익률 및 순위.6_y"]]
    merged_df.rename(columns={'수익률 및 순위_x':'1개월 수익률', 
                                '수익률 및 순위.2_x':'3개월 수익률', 
                                '수익률 및 순위.4_x':'6개월 수익률', 
                                '수익률 및 순위.6_x':'9개월 수익률', 
                                '수익률 및 순위_y':'1년 수익률', 
                                '수익률 및 순위.2_y':'2년 수익률', 
                                '수익률 및 순위.4_y':'3년 수익률', 
                                '수익률 및 순위.6_y':'5년 수익률', 
                                }, inplace=True)
    merged_df = merged_df.loc[4:]
    merged_df.reset_index(drop=True, inplace=True)
    merged_df = merged_df[~merged_df["펀드명"].str.contains("Class")]
    merged_df = merged_df[~merged_df["펀드명"].str.contains("클래스")]
    merged_df = merged_df[~merged_df["펀드명"].str.contains("종류")]
    merged_df.reset_index(drop=True, inplace=True)
    merged_df.to_excel('./excel/주식고편입형_인덱스_수익률.xls')
    print(merged_df)
    
```

```python
merge_preprocess(stock_fund_growing1, stock_fund_growing2)
```
